```python
import matplotlib.animation as animation 
import matplotlib as mpl
plt.rcParams["animation.html"] = "jshtml"

def animate(height):
    ax.set_title(f'Bz @ distance from the photospher: {dum.los_km[height].astype(int):5d} km')
    im01.set_data(dum.bz_iris[height])

fig, ax = plt.subplots()
ax.set(xlabel='Scan direction (pixel)', ylabel='Slit direction (pixel)')
im01 = ax.imshow(dum.bz_iris[0], cmap='gray', vmin=-5e2, vmax=5e2, origin='lower')
cb01 = fig.colorbar(im01)
ani = animation.FuncAnimation(fig, animate, frames=range(0, len(dum.los_km)), interval=1e2)
ani.save(filename='/sanhome/khcho/IRIS2_bfield/20190108_201920_3610108077/IRIS_FOV_bfield.mp4', 
         writer='ffmpeg', dpi=200, savefig_kwargs={'bbox_inches':'tight'})

# animation.FuncAnimation(fig, animate, frames=range(0, len(dum.los_km)), interval=1e2)
```
